### About the Dataset

**Context**:  
The growth of supermarkets in highly populated cities is increasing, and market competition is intense. **Attribute Information**:

- **Invoice ID**: Computer-generated sales slip invoice identification number.
- **Branch**: Branch of the supercenter (3 branches available, identified by A, B, and C).
- **City**: Location of the supercenters.
- **Customer Type**: Type of customer, recorded as Members (using a member card) or Normal (without a member card).
- **Gender**: Gender of the customer.
- **Product Line**: General item categorization groups, including:
  - Electronic Accessories
  - Fashion Accessories
  - Food and Beverages
  - Health and Beauty
  - Home and Lifestyle
  - Sports and Travel
- **Unit Price**: Price of each product in $.
- **Quantity**: Number of products purchased by the customer.
- **Tax**: 5% tax fee applied to the customer's purchase.
- **Total**: Total price including tax.
- **Date**: Date of purchase (Records available from January 2019 to March 2019).
- **Time**: Purchase time (10 AM to 9 PM).
- **Payment**: Payment method used by the customer (3 methods available: Cash, Credit Card, and Ewallet).
- **COGS**: Cost of Goods Sold.
- **Gross Margin Percentage**: Gross margin percentage.
- **Gross Income**: Gross income.

### Hypothesis Testing for Gender Differences in Product Preferences

To determine if there is a significant difference in the average quantity of products purchased by male and female customers, a two-sample t-test was conducted.

#### Hypotheses
- **Null Hypothesis (H0)**: The average quantity purchased by male customers is the same as the average quantity purchased by female customers.
- **Alternative Hypothesis (H1)**: The average quantity purchased by male customers is different from the average quantity purchased by female customers.

#### Test Selection
A two-sample t-test was used to compare the means of two independent groups (male and female customers) to assess if their average quantities are significantly different.

#### Results Interpretation
- **Significance Level**: 0.05
- **Decision**:
  - If the p-value obtained from the t-test is less than 0.05, the null hypothesis is rejected, indicating a significant difference in the average quantity of products purchased by male and female customers.
  - If the p-value is greater than 0.05, the null hypothesis is not rejected, indicating no significant difference in the average quantity of products purchased by male and female customers.


```python
import pandas as pd
from scipy import stats
male_data = sales_data[sales_data['Gender'] == 'Male']['Quantity']
female_data = sales_data[sales_data['Gender'] == 'Female']['Quantity']
t_statistic, p_value = stats.ttest_ind(male_data, female_data)
print(f"T-Statistic: {t_statistic}")
print(f"P-Value: {p_value}")
# Interpret the results
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: There is a significant difference in the average quantity purchased by male and female customers.")
else:
    print("Fail to reject the null hypothesis: There is no significant difference in the average quantity purchased by male and female customers.")
```

**Hypothesis Test for Gender Differences in Product Purchases Conclusion**

To investigate whether there is a significant difference in the average quantity of products purchased by male and female customers, we conducted a two-sample t-test. 
The null hypothesis (H0) stated that the average quantity purchased by male and female customers is the same. The alternative hypothesis (H1) stated that the average quantity purchased by male and female customers is different.

Using the sales data, we filtered the Quantity column by gender and performed the t-test. The results yielded a t-statistic of -2.3524 and a p-value of 0.0188. Given that the p-value is less than our significance level of 0.05, we reject the null hypothesis. 
Thus, we conclude that there is a significant difference in the average quantity purchased by male and female customers.
